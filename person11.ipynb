```
import os
from PIL import Image

fileName = os.listdir('/content/person12/person12/batch1/')
width = 160
height = 160

for img in fileName:
  pic = Image.open('/content/person12/person12/batch1/'+img)
  newpic = pic.resize((width,height),Image.ANTIALIAS)
  print(newpic)
  newpic.save('/content/new12/batch1/'+img)
```

```
import os 
from deepface import DeepFace


g = os.walk(r"/content/new11/batch1/")  
srcPath='/content/verified/'

for path,dir_list,file_list in g:  
    for file_name in file_list:  
      try:
          result  = DeepFace.verify('/content/new11/batch1/000014.png', os.path.join(path, file_name),model_name="Facenet",distance_metric='euclidean_l2')
      except:
        continue
      result  = DeepFace.verify('/content/new11/batch1/000014.png', os.path.join(path, file_name),model_name="Facenet",distance_metric='euclidean_l2')
       
      print("euclidean_l2 distance: ",result["distance"])
      print(os.path.join(path, file_name))
  
      if result["distance"]<1.1:
         print('verified:',os.path.join(path, file_name))
```
