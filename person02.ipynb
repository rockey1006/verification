```
import os
from PIL import Image

fileName = os.listdir('/content/person02/person02/batch1/')
width = 160
height = 160

for img in fileName:
  pic = Image.open('/content/person02/person02/batch1/'+img)
  newpic = pic.resize((width,height),Image.ANTIALIAS)
  print(newpic)
  newpic.save('/content/person_new02/batch1/'+img)
```

```
from deepface import DeepFace
import imageio
import imageio.core.util
def silence_imageio_warning(*args, **kwargs):
    pass
imageio.core.util._precision_warn = silence_imageio_warning
import warnings

def fxn():
    warnings.warn("deprecated", DeprecationWarning)

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    fxn()
import warnings
warnings.filterwarnings("ignore")
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3' 
import tensorflow as tf
import logging
tf.get_logger().setLevel(logging.ERROR)
```

```
import os 

g = os.walk(r"/content/person_new02/batch1/")  
#srcPath='/content/verified/'

for path,dir_list,file_list in g:  
    for file_name in file_list:  
      try:
          result  = DeepFace.verify('/content/person_new02/batch1/000005.png', os.path.join(path, file_name),model_name="Facenet",distance_metric='euclidean_l2')
      except:
        continue
      result  = DeepFace.verify('/content/person_new02/batch1/000005.png', os.path.join(path, file_name),model_name="Facenet",distance_metric='euclidean_l2')
       
      print("euclidean_l2 distance: ",result["distance"])
      print(os.path.join(path, file_name))
  
      #if result["distance"]<1.1:
       #  print('verified:',os.path.join(path, file_name))
```
