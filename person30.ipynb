```
import os 

g = os.walk(r"/content/newperson30/batch01/")  
dstPath='/content/verified13/'
if not os.path.exists(dstPath):
        os.makedirs(dstPath)

for path,dir_list,file_list in g:  
    for file_name in file_list:  
      try:
          result  = DeepFace.verify('/content/personD/needed1118_f/000007.png', os.path.join(path, file_name),model_name="Facenet",distance_metric='euclidean_l2')
      except:
        continue
      result  = DeepFace.verify('/content/personD/needed1118_f/000007.png', os.path.join(path, file_name),model_name="Facenet",distance_metric='euclidean_l2')
       
      print("euclidean_l2 distance: ",result["distance"])
      print('location:',os.path.join(path, file_name))
```
